## Axis

```python
import random
```

```python
genes = pd.read_csv("covid_filtered.tsv", sep = '\t', header = None)
```

```python
tuples = [tuple(x) for x in genes[[1,2]].to_numpy()]
```

```python
a = tuples[0:12]+tuples[14:25]+tuples[26:29]+tuples[31:]
b = tuples[12:13]+ tuples[25:26] + tuples[29:30]
c = tuples[13:14]+tuples[30:31]

intervals_from_bed = [a,b,c]
```

```python
plt.figure(figsize = (50, 5))
for i, lay in enumerate(intervals_from_bed):
    x1, x2 = zip(*lay)
    plt.hlines([i + 1] * len(x1), x1, x2, colors = sns.color_palette("rocket")[4])
    plt.vlines(x1, [i + 0.9] * len(x1), [i + 1.1] * len(x1), colors = sns.color_palette("rocket")[0])
    for b in x1:
        plt.text(b+15, i+1.1, str(genes[genes[1] == b][7].index.values[0]), fontsize = 22)
        
plt.axis([0,29768,-0.5,3.5])        
plt.savefig("axis.png")

plt.show()
```

## BG2 file format creation

```python
mut2mut
```

```python
bins = range(0,29768,100)
mut2mut['binned_1'] = pd.cut(mut2mut['pos_1'], bins)
mut2mut['binned_2'] = pd.cut(mut2mut['pos_2'], bins)
print (mut2mut)
```

```python
mut2mut_2=mut2mut.dropna(axis=0)
```

```python
mut2mut_2['start_1'] = [x.left +1 for x in mut2mut_2['binned_1']]
```

```python
mut2mut_2['stop_1'] = [x.right for x in mut2mut_2['binned_1']]

mut2mut_2['start_2'] = [x.left +1 for x in mut2mut_2['binned_2']]
mut2mut_2['stop_2'] = [x.right for x in mut2mut_2['binned_2']]
```

```python
mut2mut_2['Chr1'] = 'Chr1'
```

```python
mut2mut_2['Chr2'] = 'Chr1'
```

```python
#mut2mut_2 = mut2mut_2[mut2mut_2[4] > 5]
```

```python
mut2mut_2.groupby(['Chr1','start_1','stop_1','Chr2','start_2','stop_2'],
                  as_index = False).agg({4:'sum'}).to_csv('example.bg2', sep="\t", index=False)
```
